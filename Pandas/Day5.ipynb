Aggregate Function

```python
'''
Aggregate Function takes a set of values and returns a single result
 mean -> average
 sum -> adds up all values
 min -> smallest one
 count -> count total number of record

 unique -> Shows all unique records
 nunique -> shows unique records count

 values_count ->  shows records and their counts

'''

'''
Data -> 
  Categorica -> Nominal Ordinal
  Numerical -> Discrete Continous
'''
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Employees.csv')
df.head(10)
```

```python
#Find average salary of employees
avg_salary = df['SALARY'].mean()
print(f'Average Salary : Rs {avg_salary:.2f}')
```

```python
#Find total salary of employees
total_salary = df['SALARY'].sum()
print(f'Total Salary : Rs {total_salary}')
```

```python
#Find max salary of employees
max_salary = df['SALARY'].max()
min_salary = df['SALARY'].min()

print(f'Min Salary : Rs {min_salary}')
print(f'Max Salary : Rs {max_salary}')

count_emp = df['EMPLOYEE_ID'].count()
print(f'Employee Count : {count_emp}')

```

```python
uniq_salary = df['SALARY'].unique()
print(f'Unique salary : {uniq_salary}')


uniq_sal_count = df['SALARY'].nunique()
print('')
print(f'uniq_sal_num : {uniq_sal_count}')

uniq_sal_values = df['SALARY'].value_counts()
print('')
print(uniq_sal_values)
```

GROUPBY

```python
#Find salary total of employees by each department

Total_dep_sal = df.groupby('DEPARTMENT_ID')['SALARY'].sum().reset_index()
Total_dep_sal
```

```python
#Find total salary , average salary , max salary , min salary by job_id

job_salary_group_by_job_id = df.groupby('DEPARTMENT_ID').agg(
     Total_sal = ('SALARY','sum'),
     Avg_Sal = ('SALARY', 'mean'),
     Min_Sal = ('SALARY', 'min'),
     Max_Sal = ('SALARY', 'max')
     
     ).reset_index()

job_salary_group_by_job_id
```
