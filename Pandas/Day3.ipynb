```python
import pandas as pd
```

```python
''' 
Data cleaning:
when we have null values we have two options:

When we have numerical data -> use mean,median
When we have categorical data -> mode 


    Remove : df.dropna(axis=0/1, how='any'/'all', inplace='True'/'False') -> 0 : row and 1 : column  (by default its axis=0, how='any')
             df.drop(col_name, axis=0/1, inplace=True)

             any -> During axis=0 , if it finds any single null value in whole row it deletes it
             all -> During axis=0 , if its finds whole row null then only it deletes row

    Fill   : 
       we have mean,median,mode and custom way to fill null values


'''
```

```python
df = pd.read_csv('Employees.csv')
df.head(5)
```

```python
df.dropna() # axis = 0 and how = any ,
```

```python
df.head()
#its stil; showing commission null values becuase we didnt do inplave = true 
#Which means chanegs have been made but not saved in df 
#to save in df we need to do inplace = true


```

```python
#Dropping using inplace = True

# # df.dropna(axis=0, how='any', inplace=True)
# df.head(5)
```

```python
#Its all about drop
# Now lets go into fill 
df.head()
#Fill commission using mean
commission_mean = df['COMMISSION_PCT'].mean()
print(f"commission_mean : {commission_mean:.2f}")
```

```python
df['COMMISSION_PCT'] = df['COMMISSION_PCT'].fillna(commission_mean)
df.head(10)
```

```python
#Checking null values count in each column
df.isnull().sum()
```

```python
manager_mode = df['MANAGER_ID'].mode()
print(f"Most occuring date : {manager_mode}")
```

```python
df['MANAGER_ID'] = df['MANAGER_ID'].fillna(manager_mode)
df.head(10)
```

```python
df.isnull().sum()
```

```python
depart_mode = df['DEPARTMENT_ID'].mode()[0]
print(f"{depart_mode}")
```

```python
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].fillna(depart_mode)[0]
df.head(10)
```

```python
df.isnull().sum()
```

```python
#Manager id data is in float
#Change it to -> int

#Type casting : float -> int

df['MANAGER_ID'] = df['MANAGER_ID'].astype(int)
```
