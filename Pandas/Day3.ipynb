```python
import pandas as pd
```

```python
''' 
Data cleaning:
when we have null values we have two options:

When we have numerical data -> use mean,median
When we have categorical data -> mode 


    Remove : df.dropna(axis=0/1, how='any'/'all', inplace='True'/'False') -> 0 : row and 1 : column  (by default its axis=0, how='any')
             df.drop(col_name, axis=0/1, inplace=True)

             any -> During axis=0 , if it finds any single null value in whole row it deletes it
             all -> During axis=0 , if its finds whole row null then only it deletes row

    Fill   : 
       we have mean,median,mode and custom way to fill null values


'''
```

```python
df = pd.read_csv('Employees.csv')
df.head(5)
```

```python
df.dropna() # axis = 0 and how = any ,
```

```python
df.head()
#its stil; showing commission null values becuase we didnt do inplave = true 
#Which means chanegs have been made but not saved in df 
#to save in df we need to do inplace = true


```

```python
#Dropping using inplace = True

# # df.dropna(axis=0, how='any', inplace=True)
# df.head(5)
```

```python
#Its all about drop
# Now lets go into fill 
df.head()
#Fill commission using mean
commission_mean = df['COMMISSION_PCT'].mean()
print(f"commission_mean : {commission_mean:.2f}")
```

```python
df['COMMISSION_PCT'] = df['COMMISSION_PCT'].fillna(commission_mean)
df.head(10)
```

```python
#Checking null values count in each column
df.isnull().sum()
```

```python
manager_mode = df['MANAGER_ID'].mode()
print(f"Most occuring date : {manager_mode}")
```

```python
df['MANAGER_ID'] = df['MANAGER_ID'].fillna(manager_mode)
df.head(10)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
depart_mode = df['DEPARTMENT_ID'].mode()[0]
print(f"{depart_mode}")
```

```python
df['DEPARTMENT_ID'] = df['DEPARTMENT_ID'].fillna(depart_mode)[0]
df.head(10)
```

```python
df.isnull().sum()
```

```python
#Manager id data is in float
#Change it to -> int

#Type casting : float -> int

df['MANAGER_ID'] = df['MANAGER_ID'].astype(int)
```

```python
import matplotlib.pyplot as plt

# ----------------------------
# Scenario 1: Regular 6-year repayment
# ----------------------------
loan = 2500000
years_regular = [1, 2, 3, 4, 5, 6]
payments_regular = [552500]*6  # Annual installments in Rs.

# Scenario 2: Deferred 1-year, 6-year repayment
years_deferred = [2, 3, 4, 5, 6, 7]
payments_deferred = [598400]*6  # Annual installments in Rs.

plt.figure(figsize=(12,5))

# ----------------------------
# Draw timeline
# ----------------------------
max_year = 7
plt.hlines(0, 0, max_year+0.5, colors='black')
plt.plot(range(max_year+1), [0]*(max_year+1), 'ko')

# ----------------------------
# Loan inflow at time 0
plt.arrow(0, 0, 0, 0.8, head_width=0.08, length_includes_head=True, color='red')
plt.text(0, 0.9, f"+{loan/100000:.2f}L", ha='center', fontsize=10, color='red')
plt.text(0, 1.05, "Loan Received", ha='center', fontsize=10, color='red')

# ----------------------------
# Regular repayment arrows (outflow)
for t, p in zip(years_regular, payments_regular):
    plt.arrow(t, 0, 0, -0.6, head_width=0.08, length_includes_head=True, color='blue')
    plt.text(t, -0.75, f"-{p/100000:.2f}L", ha='center', fontsize=10, color='blue')

plt.text(6, -1.0, "Regular Repayment", ha='center', fontsize=10, color='blue')

# ----------------------------
# Deferred repayment arrows (outflow)
for t, p in zip(years_deferred, payments_deferred):
    plt.arrow(t, 0, 0, -0.6, head_width=0.08, length_includes_head=True, color='green')
    plt.text(t, -0.75, f"-{p/100000:.2f}L", ha='center', fontsize=10, color='green')

plt.text(7, -1.0, "Deferred 1-Year Repayment", ha='center', fontsize=10, color='green')

# ----------------------------
# Labels
plt.xticks(range(0,max_year+1))
plt.xlabel("Time (Years)")
plt.ylabel("Cash Flow")
plt.title("Loan Cash Flow Diagram (Inflow + Outflow)")
plt.yticks([])
plt.grid(False)
plt.show()
```
