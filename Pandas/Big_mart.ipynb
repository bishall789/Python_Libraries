Big Mart Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bigmart Sales Data.csv")
df.head(10)
```

```python
df['Item_Type'].unique()
```

```python
df.groupby('Item_Type')['Item_Weight'].agg(['min', 'max' , 'size', 'count'])
```

```python
df['Item_Weight']  = df.groupby('Item_Type')['Item_Weight'].transform(lambda x: x.fillna(x.mean()))
df.groupby('Item_Type')['Item_Weight'].agg(['min', 'max' , 'size', 'count'] )
```

```python
df.isnull().sum()
```

```python
df['Outlet_Size'].value_counts(dropna=False)
```

```python
#Fillins Outlet Type

df['Outlet_Size'] = df.groupby('Outlet_Location_Type')['Outlet_Size'].transform(lambda x: x.fillna(x.mode()[0]))
df.isnull().sum()
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Outlet_Establishment_Year'] = pd.to_datetime(df['Outlet_Establishment_Year'], format='mixed')


df['Outlet_Establishment_Year'] = df['Outlet_Establishment_Year'].dt.year
df['Outlet_Establishment_Year']
```

```python
df.info()
```

```python
df.groupby('Item_Type')['Item_Weight'].agg(['min', 'max' , 'size', 'count'])
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].str.strip()
df['Item_Identifier'] = df['Item_Identifier'].str.strip()

df['Item_Type'] = df['Item_Type'].str.strip()
df['Outlet_Identifier'] = df['Outlet_Identifier'].str.strip()
df['Outlet_Size'] = df['Outlet_Size'].str.strip()
```
