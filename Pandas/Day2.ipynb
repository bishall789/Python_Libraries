```python
import pandas as pd
```

```python
#Conditional Filtering
'''
Single Condition Filter
Using And -> & 
Using between ( used in range values only like salary)
Using Or -> | 
Using isin
'''
```

```python
df = pd.read_csv('Employees.csv')
df.head(10)
```

```python
#1. Find the employees details who works under department 90.
depart_filter = df['DEPARTMENT_ID'] == 90
df[depart_filter]
```

```python
#2. Find the details of employees whose salary is more than 5000 and less than 15000


sal_filter = (df['SALARY'] > 5000) & (df['SALARY'] < 15000)
df[sal_filter]

#Using between
sal_fil = df['SALARY'].between(5000,15000)  
df[sal_fil].head()

# When we use AND then it is exclusive 
# When we use between then it is inclusive
```

```python
#3. Find the employees details who has job id of IT_PROG or FI_ACCOUNT
job_filter = (df['JOB_ID'] == 'IT_PROG') | (df['JOB_ID'] == 'FI_ACCOUNT')
df[job_filter]

# Using isin
job_fil = df['JOB_ID'].isin(['IT_PROG','FI_ACCOUNT'])
df[job_fil].head(10)

#Why isin is more effcicient than or
```

```python
#importing dataset
df_hr = pd.read_csv("t.csv")
df_hr.head(10)
```

```python
#1. Find details of the employees with the salary range of 50000 to 70000 and who are married
#2. Details of employees who works under productiona and their employment status is active as well
#3. Details of employees who works under IT/IS department and their absences record must be more than 8

#1.
emp_sal = (df_hr['Salary'].between(50000,70000)) & (df_hr['MaritalDesc'] == 'Married')
df_hr.loc[emp_sal,['Salary','MaritalDesc']]
```

```python
#2. Details of employees who works under productiona and their employment status is active as well
emp_pr_es = (df_hr['Department'] == 'Production       ') & (df_hr['EmploymentStatus'] == 'Active')
df_hr[emp_pr_es]
```

```python
 #3. Details of employees who works under IT/IS department and their absences record must be more than 8
emp_depart_abs = (df_hr['Department'].isin(['IT/IS'])) & (df_hr['Absences'] > 8)
df_hr.loc[emp_depart_abs, ['Department','Absences']].head(10)
```
