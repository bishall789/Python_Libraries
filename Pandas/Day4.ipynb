```python
import pandas as pd
```

```python
df = pd.read_csv('t.csv')
df.head(5)
```

Clean Unwanted Data

```python
df.isnull().sum()
```

```python
'''  
These , . / and other special charater in the data -> referred as Regular Expression

# str -> Its methods 
  There are so many functions inside str method , some are : 

replace(find,replace,regex=True/False)


'''
```

```python
'''   
replace(find,replace,regex=True/False)

regex :
   True -> IF its true, comma hatxa hatna chahi tara background chahib basirahanxa , paxi ml ma effect dekauxa
   False -> If its false, comma completely hataidinxa background and onground bata ni
'''

df['Employee_Name'] = df['Employee_Name'].str.replace(',' , '' , regex=False)
df.head(10)

```

```python
'''
remove white space  using strip()
In this Department col -> Production has space
'''

df['Department'] = df['Department'].str.strip()

fil_depart = df['Department'] == 'Production'
df[fil_depart]
```

```python
df.info()
```

```python
#Chaning the dataype of date from object(string) to datatime
'''
pd.to_datetime(column, format = '%d.%m.%y' / 'mixed' )
its also datatype conversion
we used astype to convert datatype but astype can only convert python builtin dtaaype ,
     but we dont have datetime in python so we need to use such
'''

df['DateofHire'] = pd.to_datetime(df['DateofHire'], format='mixed')
df['DateofHire']
```

Data Transformation / Feature Engineering



```python
# Splitting columns
''' 
Splitting Full Name into first name , middle name and last name
split -> (find , n=index , expand=True) by default expand is false 

n gives in how many index data is split
expand tells to divide that 
 
'''

df[['First_Name' , 'Middle_Name', 'Last_Name']] = df['Employee_Name'].str.split(" ",n=2,expand=True)
df.head(10)

```

```python
#Extract year , month , day from DateofHire
df['DateofHire']
```

```python
#Extract year, month, day from date

df['year'] = df['DateofHire'].dt.year
df['month'] = df['DateofHire'].dt.month
df['day'] = df['DateofHire'].dt.day
df
```

Map

```python
df['HispanicLatino'].unique()

'''  As we can see there are 'Yes' and 'yes' , similary for no as well
we can convert whole column in upper case , lower case or proper

but its not good approach 

we should do with mapping
'''
```

```python
his_lat = {
    'no' : 'No',
    'yes' : 'Yes',
    'No' : 'No',
    'Yes' : 'Yes'
}

df['HispanicLatino'] = df['HispanicLatino'].map(his_lat)
df['HispanicLatino'].head(20)
```

```python
'''
Convert Yes -> 1 and No -> 0 to record to a new column
'''
his_m = {
    'Yes' : 1,
    'No' : 0
}

df['Hispa_Num'] = df['HispanicLatino'].map(his_m)
df['Hispa_Num'].head(20)
```
